**Dealing with missing values**

```python
train_data["fuel_type"].unique()
train_data["accident"].unique()
train_data["clean_title"].unique()
```

```python
cat_col=[x for x in train_data.columns if train_data[x].dtype=="object"]
cat_col
```

```python
num_col=[x for x in train_data.columns if train_data[x].dtype=="int"]
num_col
```

```python
num_unique_values_cat=[(x,train_data[x].nunique()) for x in cat_col]
num_unique_values_cat
```

```python
train_data["clean_title"]=train_data["clean_title"].fillna("No")
train_data["clean_title"].isnull().sum()
```

```python
train_data["fuel_type"]=train_data["fuel_type"].fillna("Unknown")
train_data["fuel_type"].isnull().sum()
```

```python
y=train_data["price"]
```

```python
X=train_data.copy()
X=X.drop(["price"],axis=1)
X.isnull().sum()
```

```python
X.drop(["id"],axis=1,inplace=True)
X
```

```python
X.drop(["model_year"],axis=1,inplace=True)
X
```

```python
test_data=pd.read_csv("/kaggle/input/usedcars/test.csv")
```

```python
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data.shape
```

```python
test_data["clean_title"]=test_data["clean_title"].fillna("No")
test_data["clean_title"].isnull().sum()
```

```python
test_data["fuel_type"]=test_data["fuel_type"].fillna("Unknown")
test_data["fuel_type"].isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
X_test=test_data.drop(["model_year"],axis=1)
X_test=test_data.drop(["id"],axis=1)
X_test
```

**Modelling and final evaluation**

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
cat=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent")),
             ("ohencoder",OneHotEncoder(handle_unknown='ignore'))])

num=Pipeline(steps=[("imputer",SimpleImputer(strategy="mean")),
            ('scaler', StandardScaler())])

processing=ColumnTransformer(transformers=[
        ('num', num, ['milage']),
        ('cat', cat, cat_col)])

from xgboost import XGBRegressor
model = XGBRegressor(random_state=0)

pipeline=Pipeline([('preprocessor', processing),
                           ('model', model)])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid =train_test_split(X,y,random_state=0,test_size=0.2,train_size=0.8)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_valid )
```

```python
from sklearn.metrics import mean_squared_error

score=mean_squared_error(y_valid,y_pred,squared=False)
score
```

**predicted scores for XGBoost**

```python
preds=pipeline.predict(X_test)
preds

submission = pd.DataFrame({
    'id': test_data["id"],
    'price': preds
})

submission.to_csv('submission.csv', index=False)
df=pd.read_csv('submission.csv')
df
```
