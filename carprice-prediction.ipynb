### Preprocessing data

```python
#so will just encode the categorical columns using ordinal encoder
categorical_col_no=[(col,data[col].nunique()) for col in data.columns if data[col].dtype=="object"]

categorical_col=[col for col in data.columns if data[col].dtype=="object"]
#will just drop Car_Name column
X=data.drop('Car_Name',axis=1)
y=data.Selling_Price

categorical_col=[col for col in X.columns if X[col].dtype=="object"]

from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.1, random_state=2)

from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder()
X_train[categorical_col]=encoder.fit_transform(X_train[categorical_col])
X_test[categorical_col]=encoder.transform(X_test[categorical_col])



```

### model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics

model1=LinearRegression()
model2=Lasso()
```

```python
model1.fit(X_train,y_train)
preds1=model1.predict(X_test)

score1=metrics.r2_score(y_test,preds1)
score1
```

```python
model2.fit(X_train,y_train)
preds2=model2.predict(X_test)

score1=metrics.r2_score(y_test,preds2)
score1
```
